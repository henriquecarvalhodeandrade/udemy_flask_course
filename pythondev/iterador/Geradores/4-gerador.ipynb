Yield
- É uma palavra chave para criar geradores. Quando você usa ele na função, você transforma ela em um gerador, que nada mais é que um tipo especial de iterador.
- Return: Quando usado, a função termina e é retornado um valor. Chamadas subsequentes iniciam a execução novamente.
- Yield: Pode ser chamada novamente e retormar a execução do ponto de parada, permitindo a produção de uma sequência de valores conforme o tempo.

```python
def numeros_impares(n):
    for i in range(1, n+1, 2):
        yield i

```

```python
for numero in numeros_impares(10):
    print(numero)
```
