```python
import itertools
```

```python
'''exemplo 1 | range()'''

for i in range(5):
    print(i)
```

```python
'''exemplo 2 | iterar lista'''

frutas = ['maça', 'banana', 'laranja']

print(frutas)

for fruta in frutas:
    print(fruta)
```

```python
'''exemplo 3 | iterar dicionario'''

pessoa = {
    'nome': 'alice',
    'idade': 30
}

for chave in pessoa:
    print(chave, pessoa[chave])

```

```python
'''exemplo 4 | enumerate()'''

for indice, fruta in enumerate(frutas):
    print(indice, fruta)
```

```python
'''exemplo 5 | zip()'''

nomes = ['Henrique','Alice','Carlos','Bob','José']
idades = [10,12,13,14,15]

for nome, idade in zip(nomes, idades):
    print(nome, idade)
```

```python
'''exemplo 6 | itertools'''

for i in itertools.count(start=1, step=2):
    if i > 20:
        break
    print(i)
```
