Enumerate
- Associates the value and index of a list of elements

```python
fruits = ['apple','grapple','orange','banana','pear']

for fruit in fruits:
    print(fruit)

# See the difference:
print()
for fruit in enumerate(fruits):
    print(fruit)

# Again:
print()
for index, fruit in enumerate(fruits):
    print(f'{index} - {fruit}')
```

```python
# One more example:

numbers = [10,20,30,40,50]

value_index = {index: value for index, value in enumerate(numbers)}
value_index
```

```python
pair_indexes = {index: value for index, value in value_index.items() if index % 2 == 0}
pair_indexes
```
