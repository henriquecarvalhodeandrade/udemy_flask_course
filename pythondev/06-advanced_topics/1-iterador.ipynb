```python
class Contador:
    def __init__(self, limite):
        '''construtor'''
        self.limite = limite
        self.atual = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.atual > self.limite:
            raise StopIteration
        else:
            valor = self.atual
            self.atual +=1
            return valor

            
```

```python
contador = Contador(10)

for number in contador:
    print(number)
```
