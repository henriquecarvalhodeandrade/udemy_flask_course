Zip
- Combines multiple iterables (list or tuple) into a single iterable

```python
names  = ["Alice","Bob","Charlie"]

ages = [18,19,20]

```

```python
# we can combine this both lists:

for name, age in zip(names, ages):
    print(f'Name: {name}, age: {age}.')  
```

```python
# affter that, now we can convert this zip into a dictionary:

people = dict(zip(names, ages))
people
```

```python
people_over_18 = {name: age for name, age, in people.items() if age > 18}
people_over_18
```
