MAP

Função que aplica uma função a todos os itens de um iterável
É uma maneira muito eficiente e concisa de transformar dados

Sem usar map:

```python
def quadrado(x):
    return x ** 2
```

```python
numeros = [1,2,3,4,5,6,7,8,9]
```

```python
for i in numeros:
    print(quadrado(i))
```

Usando Map:

```python
resultados = map(quadrado, numeros)
list(resultados)
```

```python
def celcius_para_fahrenheit(celcius):
    return (celcius * 9/5) + 32
```

```python
temperaturas_celcius = [20,30,40,50]

temperaturas = map(celcius_para_fahrenheit, temperaturas_celcius)
list(temperaturas)
```

```python
def normalizar_nome(nome):
    return nome.strip().lower()
```

```python
nomes = ["   ANA","HENRIQUE","    JOAO PAuLo ","SARa"]

nomes_normalizados = map(normalizar_nome, nomes)
list(nomes_normalizados)
```
