```python
import re
```

```python
def extrair_urls(texto):
    padrao_url = r'https://[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}'
    urls = re.findall(padrao_url, texto)

    return urls
```

```python
texto = '''
Visiste nosso site em: https://www.facebook.com/?locale=pt_BR
Insta: https://www.instagram.com
'''

resultado = extrair_urls(texto)
print(resultado)
```

```python
def validar_senha(senha):
    padrao_senha = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)[A-Za-z\d]{8,}$'
    return bool(re.match(padrao_senha, senha))
```

```python
senhas = ['senha123','senha1','Senha123','12345678','abcdefgh','Valido123','8']

print("Verificador de senhas válidas:\n")
for i in senhas:
    print(f'A senha "{i}" é válida? R: {validar_senha(i)}')
```

```python
def contar_palavars(texto):
    padrao_palavra = r'\b\w+\b'
    palavras = re.findall(padrao_palavra, texto)

    return len(palavras)
```

```python
texto_exemplo = "Esta frase possui 5 palavras!"

numero_palavras = contar_palavars(texto_exemplo)

print(numero_palavras)
```
