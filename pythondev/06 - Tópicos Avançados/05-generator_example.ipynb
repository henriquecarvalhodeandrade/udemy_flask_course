```python
'''essa função exibe as linhas que possuem o mesmo numero de palavas informadas no parâmetro.'''

def linhas_com_numero_de_palavras(arquivo, numero):
    with open(arquivo, 'r', encoding='utf-8') as file:
         for linha in file:
            if len(linha.split()) == numero:
                yield linha.strip()
```

```python
arquivo = 'texto.txt'
numero_palavars = 2

for linha in linhas_com_numero_de_palavras(arquivo, numero_palavars):
    print(linha)
```
