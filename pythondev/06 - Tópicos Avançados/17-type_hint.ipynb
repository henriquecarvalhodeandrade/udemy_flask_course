Type Hint
- A way of indicating the type of variables, parameters and function return values

```python
def my_sum(a: int, b: int) -> int:
    return a + b

my_sum(8,9)
```

```python
# One more Example

from typing import List

class Student:
    def __init__(self, name: str, grades: List[float]) -> None:
        self.name = name
        self.grades = grades

    def calculate_mean(self) -> float:
        return sum(self.grades) / len(self.grades)

def approved_students(students: List[Student], min_mean: float) -> List[str]:
    return [student.name for student in students if student.calculate_mean() >= min_mean]

students = [
    Student("Alice", [8.0,7.5,9.0]),
    Student("Bob", [10,4,6.7]),
    Student("Luna", [2.5,9.5,8.0]),
    Student("Henrique", [8.0,10.0,9.0]),
]

results = approved_students(students, 7)
results
```
